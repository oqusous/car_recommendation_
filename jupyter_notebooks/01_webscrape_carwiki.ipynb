```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
import time
from tqdm import tqdm
import pickle as pkl
```

```python
root_url ='https://www.auto-data.net/'
```

```python
brands_url = "https://www.auto-data.net/en/allbrands"
```

```python
def soups_(url):
    r = requests.get(url)
    print(r.status_code)
    
    return BeautifulSoup(r.content, 'html.parser')
```

```python
def model_url_finder(lo_fc, urls):
    dict_1 = {}
    patterns = list(map(lambda x:x.lower(),lo_fc))
    for pattern in patterns:
        for url in urls:
            if re.search(pattern, url):
                dict_1[pattern]= (url ,urls.index(url))
    return dict_1
```

```python
key_pattern = '^(?:\S+\s){1}(\S+)'
alfa_key_pattern = '^(?:\S+\s){1}(\S+\s.+\s)'
model_pattern = '^(?:\S+\s){2}(\S.+)'
alfa_model_pattern = '^(?:\S+\s){3}(\S.+)'
all_model_list={}
for brand in all_cars_list_sele:
    ks = []
    vs = []
    prices = []
    for k,v in brand.items():
        if re.findall(key_pattern, k)[0] == 'Alfa':
            ks.append(re.findall(alfa_key_pattern, k)[0])
            vs.append(re.findall(alfa_model_pattern, k)[0].lower().replace(' ','-'))
            prices.append(v)
        else:
            ks.append(re.findall(key_pattern, k)[0])
            vs.append(re.findall(model_pattern, k)[0].lower().replace(' ','-'))
            prices.append(v)
    all_model_list.update({list(set(ks))[0]: vs})
```

## Helper Functions

```python
def latest_model_url(url):
    most_recent = soups_(root_url+url).find('a', {'class':'position'}).attrs['href']
    list1 = list(map(lambda x:x.attrs['href'],soups_(root_url+most_recent).find('tbody').findAll('a')))
    return list1
```

```python
# pkl.dump([], open('test.p','wb'))
```

```python
def spec_parser(urls):
    model_list = []
    spec_dict={}
    for url in urls:
        url_full='https://www.auto-data.net/'+url
        soup_list = soups_(url_full).find('table',{"class":"cardetailsout car2"}).findAll('tr')
        spec_dict={}
        for car in soup_list:
            time.sleep(0.5)
            if car.find('th') and car.find('td'):
                try:
                    spec_dict[car.find('th').text] = car.find('td').text
                except:
                    continue
        print('Success', ':', spec_dict)
        model_list.append(spec_dict)                   
    return model_list
```

```python
def parser_looper(dict_o_cars):
    list1=[]
    print(list1)
    for k,v in dict_o_cars.items():
        url = v[0]
        print(url)
        latest_model = latest_model_url(url)
        print(latest_model)
        model_list = spec_parser(latest_model)
        print(model_list)
        list1.append(model_list)
    return list1
```
