### Links to each model

```python
def link_harvester_kbb(url1):
    url_list1 = []
    print(url_list1)
    driver.get(url1)
    driver.implicitly_wait(10)
    try:
        for b in driver.find_elements_by_css_selector('a'):
            if b.get_attribute('data-vehicleid'):
                print(b.get_attribute('href'))
                url_list1.append(b.get_attribute('href'))
            if b.get_attribute('data-analytics-pagination') and b.text == 'Next' and b.get_attribute('href'):
                print( b.get_attribute('href'))
                url_list_2 = link_harvester_kbb(b.get_attribute('href'))
                url_list1.extend(url_list_2)
    except exceptions.StaleElementReferenceException:
        pass
    return url_list1
```
